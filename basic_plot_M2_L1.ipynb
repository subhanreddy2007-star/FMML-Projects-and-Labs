 a Line Plot and a Bar Plot


**Interpretation of the Bar Plot:**

A Bar Plot (or Bar Chart) presents categorical data with rectangular bars whose lengths or heights are proportional to the values they represent. It's ideal for comparing different categories. Here, we compare the values associated with five distinct categories (A, B, C, D, E).

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']

values = [23, 45, 56, 12, 39]

plt.figure(figsize=(8, 4))

plt.bar(categories, values, color='skyblue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

 Line Plot

A Line Plot is used to display quantitative values over a continuous interval or time span. It's good for showing trends and change , example  plotted the `sin(x)` function, clearly showing its periodic nature over the given x-range.

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(0, 10, 100)

y_values = np.sin(x_values)

plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, label='sin(x) function', color='blue', linestyle='--')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.grid(True)
plt.show()
```

### Basic Box Plot Example using Matplotlib

A Box Plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can also show outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

data_category1 = np.random.normal(100, 20, 200)
data_category2 = np.random.normal(120, 15, 150)
data_category3 = np.random.normal(90, 25, 180)

1
data_category1 = np.append(data_category1, [20, 30, 180, 190])

data_to_plot = [data_category1, data_category2, data_category3]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

bp = ax.boxplot(data_to_plot, patch_artist=True, vert=True)

colors = ['#007ACC', '#FF4E50', '#5CB85C']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

for whisker in bp['whiskers']:
    whisker.set(color='#8B008B', linewidth=1.5, linestyle=':')

for cap in bp['caps']:
    cap.set(color='#8B008B', linewidth=2)

for median in bp['medians']:
    median.set(color='white', linewidth=2)

for flier in bp['fliers']:
    flier.set(marker='o', color='#E4572E', alpha=0.5)

ax.set_xticklabels(['Category 1', 'Category 2', 'Category 3'])
ax.set_ylabel('Values')
ax.set_title('Box Plot of Synthetic Data for Different Categories')
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales_data = {
    'Category': ['Total BMW Sales', 'Electric Vehicles (EVs)', 'SAVs (SUVs)', 'MINI', 'BMW Motorrad'],
    'Units Sold': [4567, 1185, 2966, 213, 1216]
}

df_sales = pd.DataFrame(sales_data)

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Units Sold', data=df_sales, palette='viridis', hue='Category', legend=False)
plt.title('BMW Group Sales Units by Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

comparison across different car models.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

data_string = """Feature Car 1 Car 2 Car 3 Car 4 Car 5
symboling 3 3 1 2 2
wheel-base 88.6 88.6 94.5 99.8 99.4
length 168.8 168.8 171.2 176.6 176.6
width 64.1 64.1 65.5 66.2 66.4
height 48.8 48.8 52.4 54.3 54.3
curb-weight 2548 2548 2823 2337 2824
engine-size 130 130 152 109 136
bore 3.47 3.47 2.68 3.19 3.19
stroke 2.68 2.68 3.47 3.40 3.40
compression-ratio 9.0 9.0 9.0 10.0 8.0
horsepower 111 111 154 102 115
peak-rpm 5000 5000 5000 5500 5500
city-mpg 21 21 19 24 18
highway-mpg 27 27 26 30 22
"""

df_raw = pd.read_csv(io.StringIO(data_string), sep='\s+')

df_raw = df_raw.set_index('Feature')

df_cars = df_raw.T
df_cars.index.name = 'Car'

print("Processed Car Data:")
display(df_cars)
```

```python
features_to_plot = ['horsepower', 'city-mpg', 'wheel-base', 'engine-size']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(features_to_plot):
    sns.barplot(x=df_cars.index, y=feature, data=df_cars, ax=axes[i], palette='viridis', hue=df_cars.index, legend=False)
    axes[i].set_title(f'{feature.replace("-", " ").title()} Comparison')
    axes[i].set_xlabel('Car Model')
    axes[i].set_ylabel(feature.replace("-", " ").title())
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

 Side Comparison of Box Plot and Violin Plot



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)

body_styles = ['Sedan', 'SUV', 'Hatchback', 'Coupe', 'Wagon']
data = {
    'body-style': np.random.choice(body_styles, 500),
    'price': np.random.normal(30000, 10000, 500)
}

df_price = pd.DataFrame(data)

df_price.loc[df_price['body-style'] == 'Sedan', 'price'] = np.random.normal(35000, 7000, len(df_price[df_price['body-style'] == 'Sedan']))
df_price.loc[df_price['body-style'] == 'SUV', 'price'] = np.random.normal(45000, 12000, len(df_price[df_price['body-style'] == 'SUV']))
df_price.loc[df_price['body-style'] == 'Hatchback', 'price'] = np.random.normal(25000, 5000, len(df_price[df_price['body-style'] == 'Hatchback']))
df_price.loc[df_price['body-style'] == 'Coupe', 'price'] = np.random.normal(40000, 10000, len(df_price[df_price['body-style'] == 'Coupe']))
df_price.loc[df_price['body-style'] == 'Wagon', 'price'] = np.random.normal(32000, 8000, len(df_price[df_price['body-style'] == 'Wagon']))

df_price['price'] = df_price['price'].apply(lambda x: max(10000, x))

print("Sample Price Data by Body-Style:")
display(df_price.head())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

sns.boxplot(x='body-style', y='price', data=df_price, palette='viridis', ax=axes[0], hue='body-style', legend=False)
axes[0].set_title('Box Plot of Price by Body-Style')
axes[0].set_xlabel('Body Style')
axes[0].set_ylabel('Price')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

sns.violinplot(x='body-style', y='price', data=df_price, palette='magma', ax=axes[1], hue='body-style', legend=False, inner='quartile')
axes[1].set_title('Violin Plot of Price by Body-Style')
axes[1].set_xlabel('Body Style')
axes[1].set_ylabel('Price')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
