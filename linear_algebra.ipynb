Multi dot methord

```python
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])   #multi dot method by the module
b=np.array([[3,1,5],[7,6,9],[6,3,8]])
c=np.array([[1,2,3],[5,4,7],[7,9,0]])
print(np.linalg.multi_dot([a,b,c]))
```

multi dot by repeated dot porduct

```python
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])
b=np.array([[3,1,5],[7,6,9],[6,3,8]])
c=np.array([[1,2,3],[5,4,7],[7,9,0]])
#print(np.linalg.multi_dot([a,b,c]))
#OR
print()
print(a.dot(b).dot(c))
```

inner poduct it is the multiplication of one matrix with other transposed matrix

```python
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])
b=np.array([[3,1,5],[7,6,9],[6,3,8]])
print(np.inner(a,b))
#OR
print()
print(np.inner(b,a))     #inter changing  the order of the matrix transposse the resulting  matrix
```

```python
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])       #Detriminents of the matrix
print(np.linalg.det(a))
```

```python
#outer product
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])
b=np.array([[3,1,5],[7,6,9],[6,3,8]])
print(np.outer(a,b))
```

```python
import numpy as np
a=np.array([[1,2,3],[5,4,7],[78,4,6]])
print(np.linalg.inv(a))
```

eigen  vector and values  represent the dierction in which they are transeformed  

```python
import numpy as np
a = np.arange(9).reshape(3,3)
eig_val, eig_vec = np.linalg.eig(a)
print('Eigenvalues are:', eig_val)
print('Eigenvectors are: ', eig_vec)
```

```python
#system of linear equation
import numpy as np

a = np.random.randint(1,34,[4,4])
b = np.array([4,8,11,8])
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```
